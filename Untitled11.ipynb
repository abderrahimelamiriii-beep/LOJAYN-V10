```python

# ==========================================================
# LOJAYN V11
# CELL 5
# SMART MONEY CONCEPTS PRO
# ==========================================================

def smart_money(df):

    # ======================================================
    # SWING HIGHS / LOWS
    # ======================================================

    lookback = 5

    df["SwingHigh"] = (
        df["High"] ==
        df["High"].rolling(lookback * 2 + 1, center=True).max()
    )

    df["SwingLow"] = (
        df["Low"] ==
        df["Low"].rolling(lookback * 2 + 1, center=True).min()
    )

    df["LastSwingHigh"] = np.nan
    df["LastSwingLow"] = np.nan

    last_high = np.nan
    last_low = np.nan

    for i in range(len(df)):

        if df["SwingHigh"].iloc[i]:
            last_high = df["High"].iloc[i]

        if df["SwingLow"].iloc[i]:
            last_low = df["Low"].iloc[i]

        df.iat[i, df.columns.get_loc("LastSwingHigh")] = last_high
        df.iat[i, df.columns.get_loc("LastSwingLow")] = last_low

    # ======================================================
    # BOS
    # ======================================================

    df["BOS"] = "NONE"

    df.loc[
        df["Close"] > df["LastSwingHigh"],
        "BOS"
    ] = "BULLISH"

    df.loc[
        df["Close"] < df["LastSwingLow"],
        "BOS"
    ] = "BEARISH"

    # ======================================================
    # CHOCH
    # ======================================================

    df["CHOCH"] = "NONE"

    for i in range(1, len(df)):

        if (
            df["BOS"].iloc[i] == "BULLISH"
            and
            df["BOS"].iloc[i-1] == "BEARISH"
        ):
            df.iat[i, df.columns.get_loc("CHOCH")] = "BULLISH"

        elif (
            df["BOS"].iloc[i] == "BEARISH"
            and
            df["BOS"].iloc[i-1] == "BULLISH"
        ):
            df.iat[i, df.columns.get_loc("CHOCH")] = "BEARISH"

    # ======================================================
    # FAIR VALUE GAP
    # ======================================================

    df["Bullish_FVG"] = (
        df["Low"].shift(-1) >
        df["High"].shift(1)
    )

    df["Bearish_FVG"] = (
        df["High"].shift(-1) <
        df["Low"].shift(1)
    )

    # ======================================================
    # ORDER BLOCKS
    # ======================================================

    df["Bullish_OB"] = (
        (df["Close"] > df["Open"]) &
        (df["Close"].shift(-1) > df["High"])
    )

    df["Bearish_OB"] = (
        (df["Close"] < df["Open"]) &
        (df["Close"].shift(-1) < df["Low"])
    )

    # ======================================================
    # LIQUIDITY SWEEP
    # ======================================================

    df["Liquidity"] = "NONE"

    df.loc[
        (df["High"] > df["High"].shift(1)) &
        (df["Close"] < df["High"].shift(1)),
        "Liquidity"
    ] = "BUY_SIDE"

    df.loc[
        (df["Low"] < df["Low"].shift(1)) &
        (df["Close"] > df["Low"].shift(1)),
        "Liquidity"
    ] = "SELL_SIDE"

    # ======================================================
    # EQUAL HIGHS / LOWS
    # ======================================================

    tolerance = 0.001

    df["Equal_High"] = (
        abs(df["High"] - df["High"].shift(1))
        / df["High"]
    ) < tolerance

    df["Equal_Low"] = (
        abs(df["Low"] - df["Low"].shift(1))
        / df["Low"]
    ) < tolerance

    # ======================================================
    # PREMIUM / DISCOUNT
    # ======================================================

    mid = (
        df["LastSwingHigh"] +
        df["LastSwingLow"]
    ) / 2

    df["Premium_Discount"] = np.where(
        df["Close"] > mid,
        "PREMIUM",
        "DISCOUNT"
    )

    return df


# ==========================================================
# APPLY SMART MONEY
# ==========================================================

btc_1h = smart_money(btc_1h)
btc_4h = smart_money(btc_4h)
btc_1d = smart_money(btc_1d)

print("=" * 70)
print("SMART MONEY READY")
print("=" * 70)

print("Swing High/Low      ✓")
print("BOS                 ✓")
print("CHOCH               ✓")
print("Fair Value Gap      ✓")
print("Order Blocks        ✓")
print("Liquidity Sweep     ✓")
print("Equal High / Low    ✓")
print("Premium / Discount  ✓")

print("=" * 70)
```

```python

# ==========================================================
# LOJAYN V11
# CELL 6
# MARKET STRUCTURE + SUPPORT/RESISTANCE + FIBONACCI
# ==========================================================

def market_structure(df):

    # ======================================================
    # TREND
    # ======================================================

    df["Trend"] = np.where(
        df["EMA50"] > df["EMA200"],
        "BULLISH",
        "BEARISH"
    )

    # ======================================================
    # SUPPORT / RESISTANCE
    # ======================================================

    df["Support"] = (
        df["Low"]
        .rolling(50)
        .min()
    )

    df["Resistance"] = (
        df["High"]
        .rolling(50)
        .max()
    )

    # ======================================================
    # DISTANCE TO SUPPORT / RESISTANCE
    # ======================================================

    support_distance = abs(df["Close"] - df["Support"])
    resistance_distance = abs(df["Resistance"] - df["Close"])

    df["Location"] = np.where(
        support_distance < resistance_distance,
        "NEAR SUPPORT",
        "NEAR RESISTANCE"
    )

    # ======================================================
    # VOLUME STATUS
    # ======================================================

    df["Volume_Status"] = np.where(
        df["Volume_Ratio"] > 1.5,
        "HIGH",
        np.where(
            df["Volume_Ratio"] < 0.8,
            "LOW",
            "NORMAL"
        )
    )

    # ======================================================
    # FIBONACCI
    # ======================================================

    highest = df["High"].rolling(200).max()
    lowest = df["Low"].rolling(200).min()

    diff = highest - lowest

    df["Fib_0"] = highest
    df["Fib_23"] = highest - diff * 0.236
    df["Fib_38"] = highest - diff * 0.382
    df["Fib_50"] = highest - diff * 0.500
    df["Fib_61"] = highest - diff * 0.618
    df["Fib_78"] = highest - diff * 0.786
    df["Fib_100"] = lowest

    # ======================================================
    # ATR STOP ZONES
    # ======================================================

    df["ATR_Stop_Buy"] = df["Close"] - df["ATR"] * 2
    df["ATR_Stop_Sell"] = df["Close"] + df["ATR"] * 2

    # ======================================================
    # EMA POSITION
    # ======================================================

    df["EMA_Position"] = np.where(
        df["Close"] > df["EMA200"],
        "ABOVE EMA200",
        "BELOW EMA200"
    )

    # ======================================================
    # TREND STRENGTH
    # ======================================================

    df["Trend_Strength"] = np.where(
        df["ADX"] >= 30,
        "VERY STRONG",
        np.where(
            df["ADX"] >= 25,
            "STRONG",
            np.where(
                df["ADX"] >= 20,
                "MODERATE",
                "WEAK"
            )
        )
    )

    # ======================================================
    # CLEAN
    # ======================================================

    df.dropna(inplace=True)

    return df


# ==========================================================
# APPLY
# ==========================================================

btc_1h = market_structure(btc_1h)
btc_4h = market_structure(btc_4h)
btc_1d = market_structure(btc_1d)

print("=" * 70)
print("MARKET STRUCTURE READY")
print("=" * 70)

print("Trend             ✓")
print("Support           ✓")
print("Resistance        ✓")
print("Fibonacci         ✓")
print("ATR Zones         ✓")
print("Volume Status     ✓")
print("Trend Strength    ✓")

print("=" * 70)
```

```python

# ==========================================================
# LOJAYN V11
# CELL 7
# PROFESSIONAL SIGNAL ENGINE + TRADE SETUP
# ==========================================================

def professional_signal(df):

    last = df.iloc[-1]

    score = 0
    reasons = []

    # ======================================================
    # EMA TREND
    # ======================================================

    if last["EMA50"] > last["EMA200"]:
        score += 20
        reasons.append("EMA Bullish")
    else:
        score -= 20
        reasons.append("EMA Bearish")

    # ======================================================
    # RSI
    # ======================================================

    if last["RSI"] >= 60:
        score += 10
        reasons.append("RSI Bullish")

    elif last["RSI"] <= 40:
        score -= 10
        reasons.append("RSI Bearish")

    # ======================================================
    # MACD
    # ======================================================

    if last["MACD"] > last["MACD_SIGNAL"]:
        score += 15
        reasons.append("MACD Bullish")
    else:
        score -= 15
        reasons.append("MACD Bearish")

    # ======================================================
    # ADX
    # ======================================================

    if last["ADX"] > 25:
        score += 5
        reasons.append("Strong Trend")

    # ======================================================
    # SMART MONEY
    # ======================================================

    if last["BOS"] == "BULLISH":
        score += 20
        reasons.append("Bullish BOS")

    elif last["BOS"] == "BEARISH":
        score -= 20
        reasons.append("Bearish BOS")

    if last["CHOCH"] == "BULLISH":
        score += 20
        reasons.append("Bullish CHOCH")

    elif last["CHOCH"] == "BEARISH":
        score -= 20
        reasons.append("Bearish CHOCH")

    if last["Bullish_OB"]:
        score += 10
        reasons.append("Bullish Order Block")

    if last["Bearish_OB"]:
        score -= 10
        reasons.append("Bearish Order Block")

    if last["Bullish_FVG"]:
        score += 10
        reasons.append("Bullish FVG")

    if last["Bearish_FVG"]:
        score -= 10
        reasons.append("Bearish FVG")

    if last["Liquidity"] == "SELL_SIDE":
        score += 10
        reasons.append("Sell Side Liquidity")

    elif last["Liquidity"] == "BUY_SIDE":
        score -= 10
        reasons.append("Buy Side Liquidity")

    # ======================================================
    # ADVANCED FILTERS
    # ======================================================

    if last["Close"] > last["Kalman"]:
        score += 5
        reasons.append("Above Kalman")
    else:
        score -= 5

    if last["Close"] > last["Wavelet"]:
        score += 5
        reasons.append("Above Wavelet")
    else:
        score -= 5

    if last["Hilbert_Trend"] == "BULLISH":
        score += 5
        reasons.append("Hilbert Bullish")
    else:
        score -= 5

    if last["Market_Phase"] == "FAST":
        score += 5

    # ======================================================
    # FINAL SIGNAL
    # ======================================================

    if score >= 50:
        signal = "STRONG BUY"

    elif score >= 25:
        signal = "BUY"

    elif score <= -50:
        signal = "STRONG SELL"

    elif score <= -25:
        signal = "SELL"

    else:
        signal = "WAIT"

    buy_probability = max(0, min(100, 50 + score))
    sell_probability = 100 - buy_probability

    return signal, score, buy_probability, sell_probability, reasons


# ==========================================================
# TRADE SETUP
# ==========================================================

def trade_setup(df):

    signal, score, buy, sell, reasons = professional_signal(df)

    entry = df["Close"].iloc[-1]
    atr = df["ATR"].iloc[-1]

    stop = None
    tp1 = None
    tp2 = None
    tp3 = None
    rr = None

    if signal in ["BUY", "STRONG BUY"]:

        stop = entry - atr * 2
        tp1 = entry + atr * 2
        tp2 = entry + atr * 3
        tp3 = entry + atr * 5

    elif signal in ["SELL", "STRONG SELL"]:

        stop = entry + atr * 2
        tp1 = entry - atr * 2
        tp2 = entry - atr * 3
        tp3 = entry - atr * 5

    if stop is not None:
        rr = abs(tp1 - entry) / abs(entry - stop)

    return (
        signal,
        score,
        buy,
        sell,
        entry,
        stop,
        tp1,
        tp2,
        tp3,
        rr,
        reasons
    )


print("=" * 70)
print("SIGNAL ENGINE READY")
print("=" * 70)
print("BUY / SELL / WAIT")
print("Probability Engine")
print("Risk Management")
print("Take Profit Generator")
print("=" * 70)
```

```python

# ==========================================================
# LOJAYN V11
# CELL 14
# PROFESSIONAL BACKTEST ENGINE
# ==========================================================

from datetime import datetime

# ==========================================================
# BACKTEST
# ==========================================================

def backtest(df):

    trades = []

    position = None

    for i in range(250, len(df)-1):

        temp = df.iloc[:i+1].copy()

        signal, score, buy, sell, entry, stop, tp1, tp2, tp3, rr, reasons = trade_setup(temp)

        next_close = df["Close"].iloc[i+1]

        if position is None:

            if signal in ["BUY", "STRONG BUY"]:

                position = {
                    "Type": "BUY",
                    "Entry": entry,
                    "SL": stop,
                    "TP": tp2,
                    "OpenTime": temp.index[-1]
                }

            elif signal in ["SELL", "STRONG SELL"]:

                position = {
                    "Type": "SELL",
                    "Entry": entry,
                    "SL": stop,
                    "TP": tp2,
                    "OpenTime": temp.index[-1]
                }

        else:

            result = None
            profit = 0

            if position["Type"] == "BUY":

                if next_close >= position["TP"]:
                    result = "WIN"
                    profit = position["TP"] - position["Entry"]

                elif next_close <= position["SL"]:
                    result = "LOSS"
                    profit = position["SL"] - position["Entry"]

            else:

                if next_close <= position["TP"]:
                    result = "WIN"
                    profit = position["Entry"] - position["TP"]

                elif next_close >= position["SL"]:
                    result = "LOSS"
                    profit = position["Entry"] - position["SL"]

            if result:

                trades.append({
                    "Open": position["OpenTime"],
                    "Close": temp.index[-1],
                    "Type": position["Type"],
                    "Result": result,
                    "Profit": round(profit,2)
                })

                position = None

    return pd.DataFrame(trades)

# ==========================================================
# RUN BACKTEST
# ==========================================================

results = backtest(btc_1h)

if len(results) > 0:

    wins = (results["Result"] == "WIN").sum()
    losses = (results["Result"] == "LOSS").sum()

    total = len(results)

    winrate = wins / total * 100

    net_profit = results["Profit"].sum()

    profit_factor = (
        results.loc[results["Profit"] > 0, "Profit"].sum() /
        abs(results.loc[results["Profit"] < 0, "Profit"].sum())
        if losses > 0 else float("inf")
    )

    print("="*70)
    print("LOJAYN V11 BACKTEST")
    print("="*70)

    print("Trades        :", total)
    print("Wins          :", wins)
    print("Losses        :", losses)
    print("Win Rate      : %.2f%%" % winrate)
    print("Net Profit    :", round(net_profit,2))
    print("Profit Factor :", round(profit_factor,2))

    print("="*70)

    results.to_csv("lojayn_v11_backtest.csv", index=False)

    telegram(f"""
📊 LOJAYN V11 BACKTEST

Trades : {total}

Wins : {wins}

Losses : {losses}

Win Rate : {winrate:.2f}%

Net Profit : {net_profit:.2f}

Profit Factor : {profit_factor:.2f}
""")

else:

    print("No completed trades found.")
```
