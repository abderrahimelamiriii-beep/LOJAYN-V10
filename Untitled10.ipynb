```python

# ==========================================================
# LOJAYN V10 - CELL 2
# DOWNLOAD MARKET DATA
# ==========================================================

SYMBOL = "BTC-USD"

def download_data(interval):

    if interval == "1h":
        period = "730d"

    elif interval == "4h":
        period = "730d"

    elif interval == "1d":
        period = "10y"

    else:
        raise ValueError("Invalid timeframe")

    df = yf.download(
        SYMBOL,
        interval=interval,
        period=period,
        auto_adjust=True,
        progress=False
    )

    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df = df.rename(columns={
        "Open":"Open",
        "High":"High",
        "Low":"Low",
        "Close":"Close",
        "Volume":"Volume"
    })

    df = df[["Open","High","Low","Close","Volume"]]

    df.dropna(inplace=True)

    return df


btc_1h = download_data("1h")
btc_4h = download_data("4h")
btc_1d = download_data("1d")

print("="*70)
print("DATA LOADED")
print("="*70)
print("1H :", len(btc_1h), "Candles")
print("4H :", len(btc_4h), "Candles")
print("1D :", len(btc_1d), "Candles")
print("="*70)
```

```python

# ==========================================================
# LOJAYN V10 - CELL 3
# PROFESSIONAL INDICATORS
# ==========================================================

def add_indicators(df):

    # ==========================
    # TREND
    # ==========================

    df["EMA20"] = ta.trend.EMAIndicator(df["Close"], window=20).ema_indicator()
    df["EMA50"] = ta.trend.EMAIndicator(df["Close"], window=50).ema_indicator()
    df["EMA100"] = ta.trend.EMAIndicator(df["Close"], window=100).ema_indicator()
    df["EMA200"] = ta.trend.EMAIndicator(df["Close"], window=200).ema_indicator()

    # ==========================
    # MOMENTUM
    # ==========================

    df["RSI"] = ta.momentum.RSIIndicator(df["Close"], window=14).rsi()

    macd = ta.trend.MACD(df["Close"])
    df["MACD"] = macd.macd()
    df["MACD_SIGNAL"] = macd.macd_signal()
    df["MACD_HIST"] = macd.macd_diff()

    df["Momentum"] = ta.momentum.ROCIndicator(
        df["Close"],
        window=10
    ).roc()

    # ==========================
    # TREND STRENGTH
    # ==========================

    adx = ta.trend.ADXIndicator(
        df["High"],
        df["Low"],
        df["Close"]
    )

    df["ADX"] = adx.adx()
    df["DI_PLUS"] = adx.adx_pos()
    df["DI_MINUS"] = adx.adx_neg()

    # ==========================
    # VOLATILITY
    # ==========================

    atr = ta.volatility.AverageTrueRange(
        df["High"],
        df["Low"],
        df["Close"]
    )

    df["ATR"] = atr.average_true_range()

    bb = ta.volatility.BollingerBands(df["Close"])

    df["BB_UPPER"] = bb.bollinger_hband()
    df["BB_MIDDLE"] = bb.bollinger_mavg()
    df["BB_LOWER"] = bb.bollinger_lband()

    # ==========================
    # VOLUME
    # ==========================

    df["VOL20"] = df["Volume"].rolling(20).mean()

    df["Volume_Ratio"] = (
        df["Volume"] /
        df["VOL20"]
    )

    df.dropna(inplace=True)

    return df


btc_1h = add_indicators(btc_1h)
btc_4h = add_indicators(btc_4h)
btc_1d = add_indicators(btc_1d)

print("="*70)
print("INDICATORS READY")
print("="*70)
```

```python

# ==========================================================
# LOJAYN V10 - CELL 4
# KALMAN + HILBERT + FOURIER + WAVELET
# ==========================================================

def advanced_analysis(df):

    # ==========================
    # KALMAN FILTER
    # ==========================

    kf = KalmanFilter(dim_x=2, dim_z=1)

    kf.x = np.array([[df["Close"].iloc[0]], [0.]])
    kf.F = np.array([[1., 1.],
                     [0., 1.]])
    kf.H = np.array([[1., 0.]])
    kf.P *= 1000.
    kf.R = 5
    kf.Q = np.array([[0.01, 0.],
                     [0., 0.01]])

    kalman = []

    for price in df["Close"]:

        kf.predict()
        kf.update(price)

        kalman.append(float(kf.x[0]))

    df["Kalman"] = kalman

    # ==========================
    # HILBERT
    # ==========================

    analytic = hilbert(df["Close"])

    phase = np.unwrap(np.angle(analytic))

    df["Hilbert_Phase"] = phase

    df["Hilbert_Trend"] = np.where(
        np.gradient(phase) > 0,
        "BULLISH",
        "BEARISH"
    )

    # ==========================
    # FOURIER
    # ==========================

    close = df["Close"].values

    fft_values = fft(close - np.mean(close))

    power = np.abs(fft_values)

    power[0] = 0

    idx = np.argmax(power[:len(power)//2])

    if idx > 0:
        cycle = len(close) / idx
    else:
        cycle = np.nan

    df["Dominant_Cycle"] = cycle
    df["Cycle_Strength"] = power[idx]

    df["Market_Phase"] = np.where(
        cycle < 40,
        "FAST",
        np.where(
            cycle < 90,
            "NORMAL",
            "SLOW"
        )
    )

    # ==========================
    # WAVELET
    # ==========================

    coeff = pywt.wavedec(
        df["Close"],
        "db4",
        level=3
    )

    coeff[1:] = [np.zeros_like(c) for c in coeff[1:]]

    wave = pywt.waverec(coeff, "db4")

    df["Wavelet"] = wave[:len(df)]

    return df


btc_1h = advanced_analysis(btc_1h)
btc_4h = advanced_analysis(btc_4h)
btc_1d = advanced_analysis(btc_1d)

print("="*70)
print("ADVANCED ANALYSIS READY")
print("="*70)
```

```python

# ==========================================================
# LOJAYN V10 - CELL 5
# SMART MONEY CONCEPTS (SMC PRO)
# ==========================================================

def smart_money(df):

    # ===============================
    # SWING HIGHS / LOWS
    # ===============================

    df["SwingHigh"] = (
        (df["High"] > df["High"].shift(1)) &
        (df["High"] > df["High"].shift(-1))
    )

    df["SwingLow"] = (
        (df["Low"] < df["Low"].shift(1)) &
        (df["Low"] < df["Low"].shift(-1))
    )

    df["LastSwingHigh"] = np.nan
    df["LastSwingLow"] = np.nan

    high = np.nan
    low = np.nan

    for i in range(len(df)):

        if df["SwingHigh"].iloc[i]:
            high = df["High"].iloc[i]

        if df["SwingLow"].iloc[i]:
            low = df["Low"].iloc[i]

        df.iloc[i, df.columns.get_loc("LastSwingHigh")] = high
        df.iloc[i, df.columns.get_loc("LastSwingLow")] = low

    # ===============================
    # BOS
    # ===============================

    df["BOS"] = "NONE"

    df.loc[
        df["Close"] > df["LastSwingHigh"],
        "BOS"
    ] = "BULLISH"

    df.loc[
        df["Close"] < df["LastSwingLow"],
        "BOS"
    ] = "BEARISH"

    # ===============================
    # CHOCH
    # ===============================

    df["CHOCH"] = "NONE"

    for i in range(2, len(df)):

        if (
            df["BOS"].iloc[i] == "BULLISH" and
            df["BOS"].iloc[i-1] == "BEARISH"
        ):
            df.iloc[i, df.columns.get_loc("CHOCH")] = "BULLISH"

        elif (
            df["BOS"].iloc[i] == "BEARISH" and
            df["BOS"].iloc[i-1] == "BULLISH"
        ):
            df.iloc[i, df.columns.get_loc("CHOCH")] = "BEARISH"

    # ===============================
    # FAIR VALUE GAP
    # ===============================

    df["Bullish_FVG"] = (
        df["Low"].shift(-1) >
        df["High"].shift(1)
    )

    df["Bearish_FVG"] = (
        df["High"].shift(-1) <
        df["Low"].shift(1)
    )

    # ===============================
    # ORDER BLOCK
    # ===============================

    df["Bullish_OB"] = (
        (df["Close"] > df["Open"]) &
        (df["Close"].shift(-1) > df["High"])
    )

    df["Bearish_OB"] = (
        (df["Close"] < df["Open"]) &
        (df["Close"].shift(-1) < df["Low"])
    )

    # ===============================
    # LIQUIDITY SWEEP
    # ===============================

    df["Liquidity"] = "NONE"

    df.loc[
        (df["High"] > df["High"].shift(1)) &
        (df["Close"] < df["High"].shift(1)),
        "Liquidity"
    ] = "BUY_SIDE"

    df.loc[
        (df["Low"] < df["Low"].shift(1)) &
        (df["Close"] > df["Low"].shift(1)),
        "Liquidity"
    ] = "SELL_SIDE"

    return df


btc_1h = smart_money(btc_1h)
btc_4h = smart_money(btc_4h)
btc_1d = smart_money(btc_1d)

print("="*70)
print("SMART MONEY READY")
print("="*70)
```

```python

# ==========================================================
# LOJAYN V10 - CELL 6
# MARKET STRUCTURE + FIBONACCI + SUPPORT/RESISTANCE
# ==========================================================

def market_structure(df):

    # ==========================
    # TREND
    # ==========================

    df["Trend"] = np.where(
        df["EMA50"] > df["EMA200"],
        "BULLISH",
        "BEARISH"
    )

    # ==========================
    # SUPPORT / RESISTANCE
    # ==========================

    df["Support"] = (
        df["Low"]
        .rolling(50)
        .min()
    )

    df["Resistance"] = (
        df["High"]
        .rolling(50)
        .max()
    )

    # ==========================
    # VOLUME STATUS
    # ==========================

    df["Volume_Status"] = np.where(
        df["Volume_Ratio"] > 1.5,
        "HIGH",
        np.where(
            df["Volume_Ratio"] < 0.8,
            "LOW",
            "NORMAL"
        )
    )

    # ==========================
    # LOCATION
    # ==========================

    dist_support = abs(df["Close"] - df["Support"])
    dist_resistance = abs(df["Resistance"] - df["Close"])

    df["Location"] = np.where(
        dist_support < dist_resistance,
        "NEAR SUPPORT",
        "NEAR RESISTANCE"
    )

    # ==========================
    # FIBONACCI
    # ==========================

    high = df["High"].rolling(200).max()
    low = df["Low"].rolling(200).min()

    df["Fib_0"] = high
    df["Fib_23"] = high - (high-low)*0.236
    df["Fib_38"] = high - (high-low)*0.382
    df["Fib_50"] = high - (high-low)*0.500
    df["Fib_61"] = high - (high-low)*0.618
    df["Fib_78"] = high - (high-low)*0.786
    df["Fib_100"] = low

    df.dropna(inplace=True)

    return df


btc_1h = market_structure(btc_1h)
btc_4h = market_structure(btc_4h)
btc_1d = market_structure(btc_1d)

print("="*70)
print("MARKET STRUCTURE READY")
print("="*70)
```

```python

# ==========================================================
# LOJAYN V10 - CELL 7
# PROFESSIONAL SIGNAL + TRADE SETUP
# ==========================================================

def professional_signal(df):

    score = 0
    reasons = []

    # ==========================
    # TREND
    # ==========================

    if df["EMA50"].iloc[-1] > df["EMA200"].iloc[-1]:
        score += 20
        reasons.append("EMA Bullish")
    else:
        score -= 20
        reasons.append("EMA Bearish")

    # ==========================
    # RSI
    # ==========================

    rsi = df["RSI"].iloc[-1]

    if rsi > 55:
        score += 10
        reasons.append("RSI Bullish")

    elif rsi < 45:
        score -= 10
        reasons.append("RSI Bearish")

    # ==========================
    # MACD
    # ==========================

    if df["MACD"].iloc[-1] > df["MACD_SIGNAL"].iloc[-1]:
        score += 15
        reasons.append("MACD Bullish")
    else:
        score -= 15
        reasons.append("MACD Bearish")

    # ==========================
    # ADX
    # ==========================

    if df["ADX"].iloc[-1] > 25:
        score += 5
        reasons.append("Strong Trend")

    # ==========================
    # SMART MONEY
    # ==========================

    if df["BOS"].iloc[-1] == "BULLISH":
        score += 20
        reasons.append("Bullish BOS")

    elif df["BOS"].iloc[-1] == "BEARISH":
        score -= 20
        reasons.append("Bearish BOS")

    if df["CHOCH"].iloc[-1] == "BULLISH":
        score += 20
        reasons.append("Bullish CHOCH")

    elif df["CHOCH"].iloc[-1] == "BEARISH":
        score -= 20
        reasons.append("Bearish CHOCH")

    if df["Bullish_OB"].iloc[-1]:
        score += 10
        reasons.append("Bullish Order Block")

    if df["Bearish_OB"].iloc[-1]:
        score -= 10
        reasons.append("Bearish Order Block")

    if df["Bullish_FVG"].iloc[-1]:
        score += 10
        reasons.append("Bullish FVG")

    if df["Bearish_FVG"].iloc[-1]:
        score -= 10
        reasons.append("Bearish FVG")

    # ==========================
    # ADVANCED FILTERS
    # ==========================

    if df["Close"].iloc[-1] > df["Kalman"].iloc[-1]:
        score += 5

    else:
        score -= 5

    if df["Close"].iloc[-1] > df["Wavelet"].iloc[-1]:
        score += 5

    else:
        score -= 5

    # ==========================
    # SIGNAL
    # ==========================

    if score >= 40:
        signal = "BUY"

    elif score <= -40:
        signal = "SELL"

    else:
        signal = "WAIT"

    buy_probability = max(0, min(100, 50 + score))
    sell_probability = 100 - buy_probability

    return signal, score, buy_probability, sell_probability, reasons


def trade_setup(df):

    signal, score, buy, sell, reasons = professional_signal(df)

    entry = df["Close"].iloc[-1]
    atr = df["ATR"].iloc[-1]

    if signal == "BUY":

        stop = entry - atr * 2
        tp1 = entry + atr * 2
        tp2 = entry + atr * 3
        tp3 = entry + atr * 4

    elif signal == "SELL":

        stop = entry + atr * 2
        tp1 = entry - atr * 2
        tp2 = entry - atr * 3
        tp3 = entry - atr * 4

    else:

        stop = tp1 = tp2 = tp3 = None

    rr = None

    if signal != "WAIT":
        rr = abs(tp1 - entry) / abs(entry - stop)

    return signal, score, buy, sell, entry, stop, tp1, tp2, tp3, rr, reasons


print("="*70)
print("SIGNAL ENGINE READY")
print("="*70)
```

```python

# ==========================================================
# LOJAYN V10 - CELL 10
# PROFESSIONAL BACKTEST
# ==========================================================

def backtest(df, initial_capital=10000):

    capital = initial_capital
    peak = capital

    equity = [capital]

    trades = 0
    wins = 0
    losses = 0

    gross_profit = 0
    gross_loss = 0

    returns = []

    for i in range(250, len(df)-1):

        signal, score, buy, sell, entry, stop, tp1, tp2, tp3, rr, _ = trade_setup(df.iloc[:i+1])

        if signal == "WAIT":
            continue

        trades += 1

        entry = df["Close"].iloc[i]
        exit_price = df["Close"].iloc[i+1]

        if signal == "BUY":
            pnl = exit_price - entry
        else:
            pnl = entry - exit_price

        capital += pnl
        equity.append(capital)

        ret = pnl / entry
        returns.append(ret)

        if pnl > 0:
            wins += 1
            gross_profit += pnl
        else:
            losses += 1
            gross_loss += abs(pnl)

        if capital > peak:
            peak = capital

    # =============================
    # STATISTICS
    # =============================

    if trades > 0:
        win_rate = wins / trades * 100
    else:
        win_rate = 0

    if gross_loss == 0:
        profit_factor = np.inf
    else:
        profit_factor = gross_profit / gross_loss

    if len(returns) > 1 and np.std(returns) != 0:
        sharpe = np.mean(returns) / np.std(returns) * np.sqrt(252)
    else:
        sharpe = 0

    equity = np.array(equity)

    running_max = np.maximum.accumulate(equity)

    drawdown = (running_max - equity) / running_max

    max_drawdown = drawdown.max() * 100

    print("="*70)
    print("BACKTEST RESULT")
    print("="*70)
    print(f"Trades         : {trades}")
    print(f"Wins           : {wins}")
    print(f"Losses         : {losses}")
    print(f"Win Rate       : {win_rate:.2f} %")
    print(f"Profit Factor  : {profit_factor:.2f}")
    print(f"Sharpe Ratio   : {sharpe:.2f}")
    print(f"Max Drawdown   : {max_drawdown:.2f} %")
    print(f"Final Capital  : {capital:.2f} $")
    print("="*70)


print("\n1H")
backtest(btc_1h)

print("\n4H")
backtest(btc_4h)

print("\n1D")
backtest(btc_1d)
```
