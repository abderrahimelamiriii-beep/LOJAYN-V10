```python

# ==========================================================
# LOJAYN AI PRO V30
# CELL 1 - DATA ENGINE
# ==========================================================

# تثبيت المكتبات
!pip -q install yfinance pandas numpy

import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np

# ==========================================================
# SETTINGS
# ==========================================================

SYMBOL = "BTC-USD"

TIMEFRAMES = {
    "1H": ("60d", "1h"),
    "4H": ("730d", "1h"),
    "1D": ("10y", "1d")
}

# ==========================================================
# DOWNLOAD FUNCTION
# ==========================================================

def load_data(period, interval):

    df = yf.download(
        SYMBOL,
        period=period,
        interval=interval,
        progress=False,
        auto_adjust=False,
        threads=False
    )

    if df.empty:
        raise Exception("No market data downloaded.")

    # إزالة MultiIndex إن وجد
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    # الاحتفاظ بالأعمدة المطلوبة
    df = df[["Open", "High", "Low", "Close", "Volume"]].copy()

    df = df.astype(float)
    df.dropna(inplace=True)

    return df

# ==========================================================
# LOAD MARKET
# ==========================================================

market = {}

# 1H
market["1H"] = load_data("60d", "1h")

# 4H من بيانات الساعة
market["4H"] = market["1H"].resample("4h").agg({
    "Open": "first",
    "High": "max",
    "Low": "min",
    "Close": "last",
    "Volume": "sum"
}).dropna()

# 1D
market["1D"] = load_data("10y", "1d")

# ==========================================================
# VALIDATION
# ==========================================================

print("=" * 70)
print("LOJAYN AI PRO V30")
print("=" * 70)

for tf in ["1H", "4H", "1D"]:

    df = market[tf]

    print(f"\nTIMEFRAME : {tf}")
    print("-" * 70)
    print("Candles :", len(df))
    print("Latest Close :", round(df["Close"].iloc[-1], 2))
    print("Latest Volume:", round(df["Volume"].iloc[-1], 2))

print("\n✅ CELL 1 LOADED SUCCESSFULLY")
```

```python

# ==========================================================
# LOJAYN AI PRO V30
# CELL 2 (PART 1)
# PROFESSIONAL INDICATOR ENGINE
# ==========================================================

analysis = {}

# ==========================================================
# EMA
# ==========================================================

def ema(series, period):
    return series.ewm(span=period, adjust=False).mean()

# ==========================================================
# RSI
# ==========================================================

def rsi(close, period=14):

    delta = close.diff()

    gain = delta.where(delta > 0, 0.0)

    loss = -delta.where(delta < 0, 0.0)

    avg_gain = gain.rolling(period).mean()

    avg_loss = loss.rolling(period).mean()

    rs = avg_gain / avg_loss.replace(0, np.nan)

    return 100 - (100 / (1 + rs))

# ==========================================================
# ATR
# ==========================================================

def atr(df, period=14):

    high = df["High"]

    low = df["Low"]

    close = df["Close"]

    tr = pd.concat([

        high - low,

        (high - close.shift()).abs(),

        (low - close.shift()).abs()

    ], axis=1).max(axis=1)

    return tr.rolling(period).mean()

# ==========================================================
# MACD
# ==========================================================

def macd(close):

    ema12 = ema(close, 12)

    ema26 = ema(close, 26)

    macd_line = ema12 - ema26

    signal = ema(macd_line, 9)

    histogram = macd_line - signal

    return macd_line, signal, histogram

# ==========================================================
# BOLLINGER BANDS
# ==========================================================

def bollinger(close, period=20):

    middle = close.rolling(period).mean()

    std = close.rolling(period).std()

    upper = middle + 2 * std

    lower = middle - 2 * std

    return upper, middle, lower
# ==========================================================
# LOJAYN AI PRO V30
# CELL 2 (PART 2)
# ADX + DI + VWAP + OBV
# ==========================================================

# ==========================================================
# VWAP
# ==========================================================

def vwap(df):

    tp = (df["High"] + df["Low"] + df["Close"]) / 3

    return (tp * df["Volume"]).cumsum() / df["Volume"].cumsum()


# ==========================================================
# OBV
# ==========================================================

def obv(df):

    obv = [0]

    close = df["Close"].values
    volume = df["Volume"].values

    for i in range(1, len(df)):

        if close[i] > close[i-1]:
            obv.append(obv[-1] + volume[i])

        elif close[i] < close[i-1]:
            obv.append(obv[-1] - volume[i])

        else:
            obv.append(obv[-1])

    return pd.Series(obv, index=df.index)


# ==========================================================
# ADX
# ==========================================================

def adx(df, period=14):

    high = df["High"]
    low = df["Low"]
    close = df["Close"]

    plus_dm = high.diff()

    minus_dm = -low.diff()

    plus_dm = plus_dm.where(
        (plus_dm > minus_dm) & (plus_dm > 0), 0.0
    )

    minus_dm = minus_dm.where(
        (minus_dm > plus_dm) & (minus_dm > 0), 0.0
    )

    tr = pd.concat([
        high - low,
        (high - close.shift()).abs(),
        (low - close.shift()).abs()
    ], axis=1).max(axis=1)

    atr_value = tr.rolling(period).mean()

    plus_di = 100 * (plus_dm.rolling(period).mean() / atr_value)

    minus_di = 100 * (minus_dm.rolling(period).mean() / atr_value)

    dx = (
        abs(plus_di - minus_di)
        / (plus_di + minus_di)
    ) * 100

    adx_line = dx.rolling(period).mean()

    return adx_line, plus_di, minus_di


# ==========================================================
# SUPPORT / RESISTANCE
# ==========================================================

def support(df, length=20):

    return df["Low"].rolling(length).min()


def resistance(df, length=20):

    return df["High"].rolling(length).max()
# ==========================================================
# LOJAYN AI PRO V30
# CELL 2 (PART 3)
# FOURIER ENGINE + FINAL ANALYSIS
# ==========================================================

def fourier_analysis(close):

    signal = np.array(close)

    signal = signal - np.mean(signal)

    fft = np.fft.rfft(signal)

    power = np.abs(fft)

    freq = np.fft.rfftfreq(len(signal), d=1)

    power[0] = 0

    idx = np.argmax(power)

    dominant_frequency = freq[idx]

    if dominant_frequency > 0:
        dominant_cycle = 1 / dominant_frequency
    else:
        dominant_cycle = 0

    cycle_strength = power[idx] / np.sum(power) * 100

    return {

        "Dominant Frequency": round(dominant_frequency,6),

        "Dominant Cycle": round(dominant_cycle,2),

        "Cycle Strength": round(cycle_strength,2),

        "Fourier Power": round(power[idx],2)

    }

# ==========================================================
# RUN ANALYSIS
# ==========================================================

for tf in market:

    df = market[tf]

    # EMA
    df["EMA20"] = ema(df["Close"],20)
    df["EMA50"] = ema(df["Close"],50)
    df["EMA200"] = ema(df["Close"],200)

    # RSI
    df["RSI"] = rsi(df["Close"])

    # ATR
    df["ATR"] = atr(df)

    # MACD
    df["MACD"], df["MACD_SIGNAL"], df["MACD_HIST"] = macd(df["Close"])

    # Bollinger
    df["BB_UPPER"], df["BB_MIDDLE"], df["BB_LOWER"] = bollinger(df["Close"])

    # VWAP
    df["VWAP"] = vwap(df)

    # OBV
    df["OBV"] = obv(df)

    # ADX
    df["ADX"], df["+DI"], df["-DI"] = adx(df)

    # Support / Resistance
    df["Support"] = support(df)
    df["Resistance"] = resistance(df)

    # Fourier
    fourier = fourier_analysis(df["Close"])

    analysis[tf] = {

        "Price": round(df["Close"].iloc[-1],2),

        "EMA20": round(df["EMA20"].iloc[-1],2),

        "EMA50": round(df["EMA50"].iloc[-1],2),

        "EMA200": round(df["EMA200"].iloc[-1],2),

        "RSI": round(df["RSI"].iloc[-1],2),

        "MACD": round(df["MACD"].iloc[-1],2),

        "ADX": round(df["ADX"].iloc[-1],2),

        "ATR": round(df["ATR"].iloc[-1],2),

        "VWAP": round(df["VWAP"].iloc[-1],2),

        "Support": round(df["Support"].iloc[-1],2),

        "Resistance": round(df["Resistance"].iloc[-1],2),

        **fourier
    }

    print("\n"+"="*70)
    print(tf)

    for k,v in analysis[tf].items():

        print(f"{k:22}: {v}")

print("\n✅ CELL 2 COMPLETED")
```

```python

# ==========================================================
# LOJAYN AI PRO V30
# CELL 3 (FIXED)
# ==========================================================

decision = {}

weights = {
    "1H": 0.20,
    "4H": 0.30,
    "1D": 0.50
}

total_score = 0
details = {}

for tf in ["1H", "4H", "1D"]:

    a = analysis[tf]

    score = 0

    # EMA
    if a["EMA20"] > a["EMA50"] > a["EMA200"]:
        score += 25
    elif a["EMA20"] < a["EMA50"] < a["EMA200"]:
        score -= 25

    # RSI
    if a["RSI"] < 30:
        score += 15
    elif a["RSI"] > 70:
        score -= 15

    # MACD
    if a["MACD"] > 0:
        score += 20
    else:
        score -= 20

    # ADX + DI (إذا كانت متوفرة)
    adx = a.get("ADX", 0)
    plus_di = a.get("+DI", None)
    minus_di = a.get("-DI", None)

    if adx > 25 and plus_di is not None and minus_di is not None:
        if plus_di > minus_di:
            score += 15
        else:
            score -= 15

    # Fourier
    if a.get("Cycle Strength", 0) > 25:
        score += 10

    # VWAP
    if a["Price"] > a["VWAP"]:
        score += 10
    else:
        score -= 10

    details[tf] = score
    total_score += score * weights[tf]

# ==========================================================
# FINAL SIGNAL
# ==========================================================

if total_score >= 60:
    signal = "🟢 STRONG BUY"
elif total_score >= 30:
    signal = "🟢 BUY"
elif total_score <= -60:
    signal = "🔴 STRONG SELL"
elif total_score <= -30:
    signal = "🔴 SELL"
else:
    signal = "🟡 WAIT"

confidence = min(abs(total_score), 100)

price = analysis["1H"]["Price"]
atr = analysis["1H"]["ATR"]

if "BUY" in signal:
    entry = price
    sl = price - atr
    tp1 = price + atr
    tp2 = price + atr * 2
    tp3 = price + atr * 3

elif "SELL" in signal:
    entry = price
    sl = price + atr
    tp1 = price - atr
    tp2 = price - atr * 2
    tp3 = price - atr * 3

else:
    entry = price
    sl = price
    tp1 = price
    tp2 = price
    tp3 = price

decision = {
    "Signal": signal,
    "Confidence": round(confidence, 1),
    "Score": round(total_score, 2),
    "Entry": round(entry, 2),
    "Stop Loss": round(sl, 2),
    "TP1": round(tp1, 2),
    "TP2": round(tp2, 2),
    "TP3": round(tp3, 2)
}

print("=" * 60)
print("LOJAYN AI PRO DECISION")
print("=" * 60)

for tf in ["1H", "4H", "1D"]:
    print(f"{tf:3} Score : {details[tf]}")

print("-" * 60)

for k, v in decision.items():
    print(f"{k:12}: {v}")

print("=" * 60)
print("✅ CELL 3 COMPLETED")
```
