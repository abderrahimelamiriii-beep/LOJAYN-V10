```python

# ==========================================================
# LOJAYN AI PRO V50
# CELL 1 - DATA ENGINE
# ==========================================================

# تثبيت المكتبات (نفذها مرة واحدة فقط)
!pip -q install yfinance numpy pandas scipy

import warnings
warnings.filterwarnings("ignore")

import yfinance as yf
import pandas as pd
import numpy as np

# ==========================================================
# SETTINGS
# ==========================================================

SYMBOL = "BTC-USD"

TIMEFRAMES = {
    "1H": ("60d", "1h"),
    "4H": ("60d", "1h"),   # سيتم تحويلها لاحقًا إلى 4 ساعات
    "1D": ("10y", "1d")
}

# ==========================================================
# DOWNLOAD FUNCTION
# ==========================================================

def load_market(period, interval):

    df = yf.download(
        tickers=SYMBOL,
        period=period,
        interval=interval,
        auto_adjust=False,
        progress=False,
        threads=False
    )

    if df.empty:
        raise Exception(f"No data for {interval}")

    # إزالة MultiIndex إذا وجد
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    # الاحتفاظ بالأعمدة المطلوبة
    df = df[["Open", "High", "Low", "Close", "Volume"]].copy()

    # تحويل إلى float
    df = df.astype(float)

    # حذف القيم الفارغة
    df.dropna(inplace=True)

    return df


# ==========================================================
# LOAD DATA
# ==========================================================

market = {}

# بيانات الساعة
market["1H"] = load_market("60d", "1h")

# إنشاء بيانات 4 ساعات من بيانات الساعة
market["4H"] = (
    market["1H"]
    .resample("4H")
    .agg({
        "Open": "first",
        "High": "max",
        "Low": "min",
        "Close": "last",
        "Volume": "sum"
    })
    .dropna()
)

# بيانات اليوم
market["1D"] = load_market("10y", "1d")

# ==========================================================
# VALIDATION
# ==========================================================

print("=" * 70)
print("LOJAYN AI PRO V50")
print("=" * 70)

for tf in ["1H", "4H", "1D"]:

    df = market[tf]

    print(f"\nTIMEFRAME : {tf}")
    print("-" * 70)
    print(f"Candles      : {len(df)}")
    print(f"Last Price   : {df['Close'].iloc[-1]:.2f}")
    print(f"Last Volume  : {df['Volume'].iloc[-1]:,.2f}")

print("\n✅ CELL 1 READY")
```

```python

# ==========================================================
# LOJAYN AI PRO V50
# CELL 2
# TREND INDICATORS ENGINE
# ==========================================================

analysis = {}

# ==========================================================
# EMA
# ==========================================================

def ema(series, period):
    return series.ewm(span=period, adjust=False).mean()

# ==========================================================
# SMA
# ==========================================================

def sma(series, period):
    return series.rolling(period).mean()

# ==========================================================
# VWAP
# ==========================================================

def vwap(df):

    tp = (df["High"] + df["Low"] + df["Close"]) / 3

    return (tp * df["Volume"]).cumsum() / df["Volume"].cumsum()

# ==========================================================
# TREND CALCULATION
# ==========================================================

for tf in market:

    df = market[tf].copy()

    # EMA
    df["EMA20"] = ema(df["Close"],20)
    df["EMA50"] = ema(df["Close"],50)
    df["EMA100"] = ema(df["Close"],100)
    df["EMA200"] = ema(df["Close"],200)

    # SMA
    df["SMA50"] = sma(df["Close"],50)
    df["SMA200"] = sma(df["Close"],200)

    # VWAP
    df["VWAP"] = vwap(df)

    market[tf] = df

    analysis[tf] = {

        "Price": round(df["Close"].iloc[-1],2),

        "EMA20": round(df["EMA20"].iloc[-1],2),

        "EMA50": round(df["EMA50"].iloc[-1],2),

        "EMA100": round(df["EMA100"].iloc[-1],2),

        "EMA200": round(df["EMA200"].iloc[-1],2),

        "SMA50": round(df["SMA50"].iloc[-1],2),

        "SMA200": round(df["SMA200"].iloc[-1],2),

        "VWAP": round(df["VWAP"].iloc[-1],2)

    }

    print("\n" + "="*70)
    print(f"{tf} TREND")

    for k,v in analysis[tf].items():
        print(f"{k:18}: {v}")

print("\n✅ CELL 2 READY")
```

```python

# ==========================================================
# LOJAYN AI PRO V50
# CELL 3
# MOMENTUM INDICATORS ENGINE
# RSI + ATR + MACD + BOLLINGER
# ==========================================================

# ==========================================================
# RSI
# ==========================================================

def rsi(close, period=14):

    delta = close.diff()

    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)

    avg_gain = gain.rolling(period).mean()
    avg_loss = loss.rolling(period).mean()

    rs = avg_gain / avg_loss.replace(0, np.nan)

    return 100 - (100 / (1 + rs))


# ==========================================================
# ATR
# ==========================================================

def atr(df, period=14):

    high = df["High"]
    low = df["Low"]
    close = df["Close"]

    tr = pd.concat([
        high - low,
        (high - close.shift()).abs(),
        (low - close.shift()).abs()
    ], axis=1).max(axis=1)

    return tr.rolling(period).mean()


# ==========================================================
# MACD
# ==========================================================

def macd(close):

    ema12 = close.ewm(span=12, adjust=False).mean()
    ema26 = close.ewm(span=26, adjust=False).mean()

    macd_line = ema12 - ema26
    signal = macd_line.ewm(span=9, adjust=False).mean()
    histogram = macd_line - signal

    return macd_line, signal, histogram


# ==========================================================
# BOLLINGER BANDS
# ==========================================================

def bollinger(close, period=20):

    middle = close.rolling(period).mean()

    std = close.rolling(period).std()

    upper = middle + (2 * std)
    lower = middle - (2 * std)

    return upper, middle, lower


# ==========================================================
# CALCULATE
# ==========================================================

for tf in market:

    df = market[tf]

    df["RSI"] = rsi(df["Close"])

    df["ATR"] = atr(df)

    (
        df["MACD"],
        df["MACD_SIGNAL"],
        df["MACD_HIST"]
    ) = macd(df["Close"])

    (
        df["BB_UPPER"],
        df["BB_MIDDLE"],
        df["BB_LOWER"]
    ) = bollinger(df["Close"])

    market[tf] = df

    analysis[tf].update({

        "RSI": round(df["RSI"].iloc[-1],2),

        "ATR": round(df["ATR"].iloc[-1],2),

        "MACD": round(df["MACD"].iloc[-1],2),

        "MACD Signal": round(df["MACD_SIGNAL"].iloc[-1],2),

        "MACD Histogram": round(df["MACD_HIST"].iloc[-1],2),

        "BB Upper": round(df["BB_UPPER"].iloc[-1],2),

        "BB Middle": round(df["BB_MIDDLE"].iloc[-1],2),

        "BB Lower": round(df["BB_LOWER"].iloc[-1],2)

    })

    print("\n" + "="*70)
    print(f"{tf} MOMENTUM")

    for k, v in analysis[tf].items():
        print(f"{k:20}: {v}")

print("\n✅ CELL 3 READY")
```

```python

# ==========================================================
# LOJAYN AI PRO V50
# CELL 4
# TREND STRENGTH + VOLUME
# ADX + DI + OBV + MFI + CMF
# ==========================================================

# ==========================================================
# OBV
# ==========================================================

def obv(df):

    values = [0]

    close = df["Close"].values
    volume = df["Volume"].values

    for i in range(1, len(df)):

        if close[i] > close[i-1]:
            values.append(values[-1] + volume[i])

        elif close[i] < close[i-1]:
            values.append(values[-1] - volume[i])

        else:
            values.append(values[-1])

    return pd.Series(values, index=df.index)


# ==========================================================
# MFI
# ==========================================================

def mfi(df, period=14):

    tp = (df["High"] + df["Low"] + df["Close"]) / 3

    mf = tp * df["Volume"]

    direction = tp.diff()

    positive = mf.where(direction > 0, 0).rolling(period).sum()

    negative = mf.where(direction < 0, 0).rolling(period).sum()

    ratio = positive / negative.replace(0, np.nan)

    return 100 - (100 / (1 + ratio))


# ==========================================================
# CMF
# ==========================================================

def cmf(df, period=20):

    mfm = (
        ((df["Close"] - df["Low"]) -
         (df["High"] - df["Close"]))
        / (df["High"] - df["Low"]).replace(0, np.nan)
    )

    mfv = mfm * df["Volume"]

    return (
        mfv.rolling(period).sum()
        /
        df["Volume"].rolling(period).sum()
    )


# ==========================================================
# ADX
# ==========================================================

def adx(df, period=14):

    high = df["High"]
    low = df["Low"]
    close = df["Close"]

    plus_dm = high.diff()

    minus_dm = -low.diff()

    plus_dm = plus_dm.where(
        (plus_dm > minus_dm) & (plus_dm > 0),
        0
    )

    minus_dm = minus_dm.where(
        (minus_dm > plus_dm) & (minus_dm > 0),
        0
    )

    tr = pd.concat([
        high - low,
        (high - close.shift()).abs(),
        (low - close.shift()).abs()
    ], axis=1).max(axis=1)

    atr_value = tr.rolling(period).mean()

    plus_di = 100 * (
        plus_dm.rolling(period).mean() / atr_value
    )

    minus_di = 100 * (
        minus_dm.rolling(period).mean() / atr_value
    )

    dx = (
        (plus_di - minus_di).abs()
        /
        (plus_di + minus_di).replace(0, np.nan)
    ) * 100

    adx_line = dx.rolling(period).mean()

    return adx_line, plus_di, minus_di


# ==========================================================
# CALCULATE
# ==========================================================

for tf in market:

    df = market[tf]

    df["OBV"] = obv(df)

    df["MFI"] = mfi(df)

    df["CMF"] = cmf(df)

    (
        df["ADX"],
        df["+DI"],
        df["-DI"]
    ) = adx(df)

    market[tf] = df

    analysis[tf].update({

        "ADX": round(df["ADX"].iloc[-1],2),

        "+DI": round(df["+DI"].iloc[-1],2),

        "-DI": round(df["-DI"].iloc[-1],2),

        "OBV": round(df["OBV"].iloc[-1],2),

        "MFI": round(df["MFI"].iloc[-1],2),

        "CMF": round(df["CMF"].iloc[-1],4)

    })

    print("\n" + "="*70)
    print(f"{tf} STRENGTH")

    for k,v in analysis[tf].items():
        print(f"{k:20}: {v}")

print("\n✅ CELL 4 READY")
```

```python

# ==========================================================
# LOJAYN AI PRO V50
# CELL 5
# FOURIER + HILBERT + SUPPORT/RESISTANCE
# ==========================================================

# ==========================================================
# SUPPORT / RESISTANCE
# ==========================================================

def support(df, length=20):
    return df["Low"].rolling(length).min()


def resistance(df, length=20):
    return df["High"].rolling(length).max()


# ==========================================================
# FOURIER ANALYSIS
# ==========================================================

def fourier_analysis(close):

    signal = np.asarray(close, dtype=float)

    signal = signal - np.mean(signal)

    fft = np.fft.rfft(signal)

    power = np.abs(fft)

    freq = np.fft.rfftfreq(len(signal), d=1)

    power[0] = 0

    idx = np.argmax(power)

    dominant_frequency = freq[idx]

    dominant_cycle = (
        1 / dominant_frequency
        if dominant_frequency > 0
        else 0
    )

    cycle_strength = (
        power[idx] / power.sum() * 100
        if power.sum() > 0
        else 0
    )

    return {
        "Dominant Frequency": round(dominant_frequency, 6),
        "Dominant Cycle": round(dominant_cycle, 2),
        "Cycle Strength": round(cycle_strength, 2),
        "Fourier Power": round(power[idx], 2)
    }


# ==========================================================
# HILBERT PHASE (Approximation)
# ==========================================================

def hilbert_phase(close):

    x = np.asarray(close, dtype=float)

    dx = np.gradient(x)

    phase = np.degrees(
        np.arctan2(dx[-1], x[-1] - np.mean(x))
    )

    if phase < 0:
        phase += 360

    if phase < 90:
        direction = "Rising"

    elif phase < 180:
        direction = "Top"

    elif phase < 270:
        direction = "Falling"

    else:
        direction = "Bottom"

    return phase, direction


# ==========================================================
# FOURIER FORECAST
# ==========================================================

def fourier_forecast(close, steps=20):

    signal = np.asarray(close, dtype=float)

    signal = signal - signal.mean()

    fft = np.fft.rfft(signal)

    freq = np.fft.rfftfreq(len(signal))

    power = np.abs(fft)

    power[0] = 0

    idx = np.argmax(power)

    if freq[idx] == 0:
        return [close.iloc[-1]] * steps

    amplitude = power[idx] / len(signal)

    phase = np.angle(fft[idx])

    t = np.arange(len(signal), len(signal) + steps)

    forecast = (
        amplitude
        * np.cos(2 * np.pi * freq[idx] * t + phase)
        + close.mean()
    )

    return forecast.tolist()


# ==========================================================
# CALCULATE
# ==========================================================

for tf in market:

    df = market[tf]

    df["Support"] = support(df)

    df["Resistance"] = resistance(df)

    phase, direction = hilbert_phase(df["Close"])

    forecast = fourier_forecast(df["Close"], 20)

    fourier = fourier_analysis(df["Close"])

    market[tf] = df

    analysis[tf].update({

        "Support": round(df["Support"].iloc[-1], 2),

        "Resistance": round(df["Resistance"].iloc[-1], 2),

        "Hilbert Phase": round(phase, 2),

        "Cycle Direction": direction,

        "Forecast 5": round(forecast[4], 2),

        "Forecast 10": round(forecast[9], 2),

        "Forecast 20": round(forecast[19], 2),

        **fourier

    })

    print("\n" + "=" * 70)
    print(f"{tf} CYCLE ANALYSIS")

    for k, v in analysis[tf].items():
        print(f"{k:22}: {v}")

print("\n✅ CELL 5 READY")
```

```python

# ==========================================================
# LOJAYN AI PRO V50
# CELL 6
# AI DECISION ENGINE
# ==========================================================

signals = {}

for tf in market:

    a = analysis[tf]

    score = 0

    # ======================================================
    # EMA TREND
    # ======================================================

    if a["Price"] > a["EMA20"]:
        score += 5

    if a["EMA20"] > a["EMA50"]:
        score += 10

    if a["EMA50"] > a["EMA100"]:
        score += 10

    if a["EMA100"] > a["EMA200"]:
        score += 15

    # ======================================================
    # RSI
    # ======================================================

    if 45 <= a["RSI"] <= 70:
        score += 10

    elif a["RSI"] < 30:
        score += 15

    elif a["RSI"] > 70:
        score -= 10

    # ======================================================
    # MACD
    # ======================================================

    if a["MACD"] > a["MACD Signal"]:
        score += 10

    if a["MACD Histogram"] > 0:
        score += 5

    # ======================================================
    # ADX + DI
    # ======================================================

    if a["ADX"] > 25:

        if a["+DI"] > a["-DI"]:
            score += 15
        else:
            score -= 15

    # ======================================================
    # VWAP
    # ======================================================

    if a["Price"] > a["VWAP"]:
        score += 5

    # ======================================================
    # FOURIER
    # ======================================================

    if a["Cycle Direction"] == "Rising":
        score += 10

    elif a["Cycle Direction"] == "Falling":
        score -= 10

    # ======================================================
    # FORECAST
    # ======================================================

    if a["Forecast 5"] > a["Price"]:
        score += 5

    if a["Forecast 10"] > a["Price"]:
        score += 5

    if a["Forecast 20"] > a["Price"]:
        score += 5

    # ======================================================
    # FINAL SIGNAL
    # ======================================================

    if score >= 70:
        signal = "STRONG BUY"

    elif score >= 50:
        signal = "BUY"

    elif score >= 35:
        signal = "WAIT"

    elif score >= 20:
        signal = "SELL"

    else:
        signal = "STRONG SELL"

    confidence = max(0, min(score, 100))

    signals[tf] = {

        "Signal": signal,

        "Score": score,

        "Confidence": confidence

    }

# ==========================================================
# DISPLAY
# ==========================================================

print("=" * 70)
print("AI DECISION ENGINE")
print("=" * 70)

for tf in signals:

    print(f"\n{tf}")

    print("-" * 70)

    print(f"Signal      : {signals[tf]['Signal']}")

    print(f"Score       : {signals[tf]['Score']}")

    print(f"Confidence  : {signals[tf]['Confidence']}%")

print("\n✅ CELL 6 READY")
```

```python

# ==========================================================
# LOJAYN AI PRO V50
# CELL 7
# TRADE SETUP + RISK MANAGEMENT
# ==========================================================

trade_setup = {}

for tf in market:

    a = analysis[tf]
    s = signals[tf]

    price = a["Price"]
    atr = a["ATR"]

    support = a["Support"]
    resistance = a["Resistance"]

    signal = s["Signal"]

    # ======================================================
    # BUY
    # ======================================================

    if signal in ["BUY", "STRONG BUY"]:

        entry = price

        stop_loss = max(
            support,
            price - (1.5 * atr)
        )

        tp1 = price + (1.5 * atr)
        tp2 = price + (3 * atr)
        tp3 = resistance

    # ======================================================
    # SELL
    # ======================================================

    elif signal in ["SELL", "STRONG SELL"]:

        entry = price

        stop_loss = min(
            resistance,
            price + (1.5 * atr)
        )

        tp1 = price - (1.5 * atr)
        tp2 = price - (3 * atr)
        tp3 = support

    # ======================================================
    # WAIT
    # ======================================================

    else:

        entry = price

        stop_loss = np.nan
        tp1 = np.nan
        tp2 = np.nan
        tp3 = np.nan

    # ======================================================
    # RISK
    # ======================================================

    if a["ADX"] >= 35:
        risk = "LOW"

    elif a["ADX"] >= 20:
        risk = "MEDIUM"

    else:
        risk = "HIGH"

    trade_setup[tf] = {

        "Entry": round(entry, 2),

        "Stop Loss": (
            round(stop_loss, 2)
            if pd.notna(stop_loss)
            else "-"
        ),

        "Take Profit 1": (
            round(tp1, 2)
            if pd.notna(tp1)
            else "-"
        ),

        "Take Profit 2": (
            round(tp2, 2)
            if pd.notna(tp2)
            else "-"
        ),

        "Take Profit 3": (
            round(tp3, 2)
            if pd.notna(tp3)
            else "-"
        ),

        "Risk": risk

    }

# ==========================================================
# DISPLAY
# ==========================================================

print("=" * 70)
print("TRADE SETUP")
print("=" * 70)

for tf in trade_setup:

    print(f"\n{tf}")
    print("-" * 70)

    for k, v in trade_setup[tf].items():
        print(f"{k:18}: {v}")

print("\n✅ CELL 7 READY")
```
